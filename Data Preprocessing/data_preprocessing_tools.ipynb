```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Features: parametri indipendenti

Dependent variables: variabile di cui si vuole fare la previsione in funzione dei parametri indipendenti

```python
dataset = pd.read_csv("Data.csv")
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
print(x)
print(y)
```

Se ho un grande dataset, posso cancellare le righe che hanno dati mancanti

Se ho un dataset piccolo, posso sostituire i valori vuoti con la media dei valori della colonna (o mediana, max, min ecc.)

Per sicurezza, includere sul fit() tutte le colonne numeriche, anche se non si vedono valori nulli. Potrebbero esserci ma il dataset potrebbe essere troppo grande per notarli.

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy="mean")
imputer.fit(x[:,1:3])
x[:,1:3] = imputer.transform(x[:,1:3])
print(x)
```

Anche i valori in stringa vanno codificati in numero.

I booleani possono diventare 0 e 1

Le stringhe possono diventare numeri ma questo darebbe un ordine ad essi che potrebbe essere mal interpretati. Meglio codificarli in vettori.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
print(x)
```

```python

from sklearn.preprocessing import LabelEncoder
y = LabelEncoder().fit_transform(y)
print(y)
```

Quando fare il feature scaling? Sempre dopo aver diviso il test set dal traning set.

Il test set deve sembrare un set completamente nuovo, come dei dati nuovi dalla produzione. Se faccio lo scaling prima della suddivisione, avrò un leak di informazioni sul training set dal set di test, che nella realtà non avrei.

E' consigliato avere l'80% dei dati come training set e il 20% come test.

random_state evita di selezionare i dati in modo random, per avere un risultato sempre uguale (solo per consistenza con il corso).

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

Il feature scaling serve per evitare che dati con ordini di grandezza molto diversi non vadano a sopperire dati più piccoli, che potrebbero non essere considerati anche se importanti. Non è sempre necessario.

Normalizzazione: funziona bene se le feature seguono una distribuzione normale.

Standardizzazione: funziona bene sempre.

Si fa lo scaling su test e training separatamente. Il fit viene fatto solo sul training, e poi applicato sia a test che a traning separatamente.

Non bisogna mai applicare lo scaling alle variabili "dummy" (variabili di testo codificate in numeri).

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,3:] = sc.fit_transform(X_train[:,3:])
X_test[:,3:] = sc.transform(X_test[:,3:])
print(X_train)
print(X_test)
```
