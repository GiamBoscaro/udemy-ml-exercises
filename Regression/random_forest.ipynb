```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

Vengono selezionati n subset casuali. Per ogni subset viene generato un decision tree regression. La predizione viene fatta su tutti gli n decision tree e i risultati vengono mediati.

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100)
regressor.fit(X, y)
```

```python
y_pred = regressor.predict([[6.5]])
print(y_pred)
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color="blue")
plt.title("Truth or Bluff (Random Forest)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```
