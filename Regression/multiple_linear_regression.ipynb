```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('50_Startups.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Non è necessario fare feature scaling, gli ordini di grandezza differenti vengono assorbiti dai coefficienti del polinomio.

Non è necessario. verificare se la relazione tra i dati è lineare prima di provare questo modello. Si perde più tempo a testare la linearità tra le variabili che a produrre il modello.

Se la relazione non è lineare, semplicemente il modello avrà un'accuratezza minore che un altro modello.

La classe utilizzata per creare la regressione si preoccupa di eliminare le dummy variable dipendenti tra loro. Inoltre gestisce in automatico la selezione delle variabili indipendenti e l'eliminazione di quelle dipendenti tra di loro (seleziona le feature che hanno più importanza statisticamente)

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
delta = y_pred-y_test
delta_n = (y_pred-y_test)/y_test*100
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1), delta.reshape(len(delta),1), delta_n.reshape(len(delta_n),1) ), axis=1))
```

```python
print(regressor.predict([[1, 0, 0, 160000, 130000, 300000]]))
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```
